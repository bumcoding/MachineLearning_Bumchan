```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
# 회귀 그래프(regression plot) : 회귀식을 적용하여
# 선형회귀 추세선을 그래프에 함께 작성
# 선형회귀 추세선 : 데이터를 기반으로 데이터의 x값 대비 y값 변화를 예측하는 직선
# 함수 regplot 사용
```

```python
sns.regplot(x="total_bill", y="tip", data=tips, x_ci=95)
# 매개변수 x_ci는 신뢰구간의 비율을 나타냄
# x_ci = 95 (상위 2.5, 하위2.5 구간을 제외하고 출력함), 대부분 연속형 데이터...
```

```python
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
```

```python
tips = sns.load_dataset("tips")
sns.countplot(x="smoker", hue="time", data=tips)
```

```python
sns.barplot(x="day", y="total_bill", data=tips)
# barplot = y 값이 연속형 값일 경우 해당 값들의 평균을 나타냄
#데이터의 신뢰구간을 검은색 막대로 표현 
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker",data=tips, palette="muted")
# violinplot = 상자그림과 분포도를 한 번에 나타낼 수 있음
# x축에는 범주형 데이터, y축에는 연속형 데이터
```

```python
g = sns.FacetGrid(tips, col="time", row="sex") 
g.map(sns.scatterplot, "total_bill", "tip") # (그래프 종류, x축, y축)
# FacetGrid = 그래프의 틀만 제공하여 적당한 그래프를 그려주는 클래스
```

```python
g = sns.FacetGrid(tips, col="time", row="sex")
g.map_dataframe(sns.histplot, x="total_bill")
# histplot은 누적합산하여 _dataframe을 가져온다
```

```python
import plotly.express as px

df = px.data.iris() # iris는 판다스 데이터프레임
df # 데이터 파악하기
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length")
fig.show()

# iris 데이터셋을 호출하여 간단한 그래프를 생성
# 래퍼 모듈인 express를 호출한 뒤 산점도를 호출
```
