```python
import matplotlib.pyplot as plt # matplotlib 모듈 호출
X = range(100) # 0~99
Y = range(100) # 0~99
plt.plot(X, Y) # 그래프를 만듦
```

```python
X = range(100) #0~99
Y = range(100) #0~99
plt.plot(X, Y)
plt.show() # 그래프를 출력
```

```python
import numpy as np #numpy 모듈 호출
X_1 = range(100)
Y_1 = [np.cos(value) for value in X] 
X_2 = range(100)
Y_2 = [np.sin(value) for value in X]
plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig, ax = plt.subplots() # (1) figure와 axes 객체 할당
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]
ax.plot(X_1, Y_1) # (2) plot 함수를 사용하여 그래프 생성
ax.set(title='cos graph', # (3)그래프 제목,X축 라벨,Y축 라벨 설정
xlabel='X', 
ylabel='Y')
plt.show() # (4) 그래프 출력
```

```python
fig = plt.figure() # (1) figure 반환
fig.set_size_inches(10,10) # (2) figure의 크기 지정
ax_1 = fig.add_subplot(1,2,1) # (3) 첫 번째 그래프 생성 (행,열,1)
ax_2 = fig.add_subplot(1,2,2) # (4) 두 번째 그래프 생성 (행,열,2)
ax_1.plot(X_1, Y_1, c="b") # (5) 첫 번째 그래프 설정
ax_2.plot(X_2, Y_2, c="#FFAAFF") # (6) (색은 16진수컬러로 바꾸기 가능)
plt.show() # (7) 그래프 출력
```

```python
# (1) x 값과 y_n 값 생성
x = np.linspace(-1,1,100) #-1~1까지 100개 숫자로 분할

y_1 = np.sin(x) # sin함수
y_2 = np.cos(x) # cos함수
y_3 = np.tan(x) # tan함수
y_4 = np.exp(x) # 지수함수(exponential)

fig, ax = plt.subplots(2, 2) # (2)2행 2열 figure 객체를 생성

ax[0, 0].plot(x, y_1) # (3) 첫 번째 그래프 생성
ax[0, 1].plot(x, y_2) # (4) 두 번째 그래프 생성
ax[1, 0].plot(x, y_3) # (5) 세 번째 그래프 생성
ax[1, 1].plot(x, y_4) # (6) 네 번째 그래프 생성

plt.show()
```

```python
 N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N) # 색깔을 랜덤하게 부여
area = np.pi * (15 * np.random.rand(N))**2 #원 넓이
plt.scatter(x, y, s=area, c=colors, alpha=0.5)# 산점도 커스텀마이징
# s = 원의 넓이, c = 색깔, alpha = 투명도 
plt.show()
```

```python
# (1) 데이터 생성
data = [[5., 25., 50., 20.], # 파이썬에서 기본적으로 제공하는 데이터
[4., 23., 51., 17],
[6., 22., 52., 19]]

# (2) X 좌표 시작점
X = np.arange(0,8,2) #0~7까지 2간격으로 배열 생성 [0,2,4,6]

# (3) 3개의 막대그래프 생성 X+막대크기
plt.bar(X + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)

# (4) X축에 표시될 이름과 위치 설정
plt.xticks(X+0.50, ("A","B","C","D")) # 0.5로 그래프 가운데 정렬

# (5) 막대그래프 출력
plt.show()
```

```python
# 누적 막대그래프

data = np.array([[5., 25., 50., 20.], # 넘파이 array에서 제공
                [4., 23., 51., 17],
                [6., 22., 52., 19]])

print(data)
data.shape
```

```python
color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1]) # data.shape=>[3,4]
#np.arange(4) => [0,1,2,3]

for i in range(3):
    plt.bar(X, data[i],
        bottom = np.sum(data[:i], axis=0),
        color = color_list[i],
        label=data_label[i])

plt.legend()
plt.show()
```

```python
# 히스토그램(데이터의 분포 표현)

N = 1000 # 히스토그램 총 개수

X = np.random.normal(size=N) #정규분포로 램덤한 숫자 추출
plt.hist(X,bins=100) # hist = 히스토그램 생성, bins = 막대 개수

plt.show()
```
