```python
import pandas as pd
import numpy as np

raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'], 
'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'], 
'age': [42, np.nan, 36, 24, 73], 
'sex': ['m', np.nan, 'f', 'm', 'f'], 
'preTestScore': [4, np.nan, np.nan, 2, 3],
'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])
df
```

```python
df.isnull() # NaN 값이 존재할 경우 True
```

```python
df.dropna() # 결측치가 나온 열이나 행을 삭제
```

```python
df_cleaned = df.dropna(how='all')
df_cleaned
```

```python
df_cleaned = df.dropna(how='any')
df_cleaned
```

```python
df['location'] = np.nan
df.dropna(axis=1, how='all')
```

```python
df.dropna(axis=0, thresh=1)
```

```python
df.dropna(thresh=5)
```

```python
df.fillna(0) # 비어있는 값들을 괄호 안의 값으로 채우기
```

```python
df["preTestScore"].fillna(df["preTestScore"].mean(), inplace=True)
df
# .mean() 평균값에 넣어준다, inplace는 변수 자체의 값을 변경해주라는 명령어
```

```python
df.loc[1,"sex"]="m" # 성별없는 부분 임의로 채워주고 진행함
df
```

```python
edges = pd.DataFrame({'source': [0, 1, 2], 
                      'target': [2, 2, 3],
                      'weight': [3, 4, 5],
                      'color': ['red', 'blue', 'blue']})
edges # 데이터 프라임 생성
```

```python
edges.dtypes # 타입 확인
```

```python
pd.get_dummies(edges) # 범주형 데이터 color에 가변수 추가
```

```python
pd.get_dummies(edges["color"]) # color 값만 따로 추출
```

```python
weight_dict = {3:"M", 4:"L", 5:"XL"}
edges["weight_sign"] = edges["weight"].map(weight_dict)
edges
```

```python
weight_sign = pd.get_dummies(edges["weight_sign"]) # 분리해서 만들어주는 부분
weight_sign
```

```python
edges = pd.concat([edges, weight_sign], axis=1)
# concat으로 데이터 두개를 합쳐준다, axis=1 열로 붙여라
edges
```

```python
 raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['regiment', 'company', 'name', 'preTestScore', 'postTestScore'])
df
```

```python
bins = [0, 25, 50, 75, 100] # bins 정의(0-25, 25-50, 60-75, 75-100) 구간별로 나눔
group_names = ['Low', 'Okay', 'Good', 'Great']
categories = pd.cut(df['postTestScore'], bins, labels=group_names)
# pd.cut(post데이터를, bins기준으로 나누겠다, labels을 group_names로)
categories
```

```python
df = pd.DataFrame(
{'A':[14.00, 90.20, 90.95, 96.27, 91.21],
'B':[103.02, 107.26, 110.35, 114.23, 114.68], 
'C':['big','small','big','small','small']})
df
```

```python
(df["A"] - df["A"].min() ) / (df["A"].max() -df["A"].min())
# 최솟값-최댓값 정규화 수식
```

```python
df["A"].min() # 최솟값(상수값)
```

```python
df["A"].max() # 최댓값(상수값)
```

```python
(df["B"] - df["B"].mean() ) / (df["B"].std())
# z-스코어 정규화 수식 
```
